## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline # if you write this line then you don't need to write plt.show()
```

```python
x =np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
y =x*x
```

```python
plt.plot(x,y,'r*--',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Diagram')
```

```python
print(x)
print(y)
```

```python
plt.subplot(2,2,1)# 1 is the first fig position
plt.plot(x,y,'r')
plt.subplot(2,2,2) # 2 is the second fig position
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
np.pi
```

```python
x =np.arange(0,4*np.pi,0.1)
y =np.sin(x)
plt.title('sine wave form')
plt.plot(x,y)
#plt.show()
```

```python
labels ='Python','C++','Ruby','Java'
sizes =[215,130,245,210]
colors =['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0)# explode 1st slice
plt.pie(sizes,explode=explode,labels =labels,colors =colors,
       autopct='%1.1f%%',shadow =True)# 1.1.f is in floating format
plt.axis('equal')
#plt.show()
```

## Seaborn tutorial

### Distribution plots
* distplot
* jointplot
* pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation with heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very importatnt in feature selection.

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joint plot
 A joint plot allows to study the relationship between 2 numeric variables. The central chart dislay their correlation. It is usually a scatter plot, a hexbin plot, a 2D histogram or a 2D density plot.

## Univariate Analysis

```python
sns.jointplot(x='tip',y ='total_bill',data =df,kind ='hex')
```

## Pair plot
A 'pair plot' is also knowna as scatterlot, in which one variable in the same data row is matched with another variable's value, like this : Pair plots are just elaborations on this, showing all variables paired with all varibles paired with all the other variables 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

## Dist plot
This plot helps to check the distribution of the column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde =False,bins=10)
```

## Categorical plots
* Boxplot

```python
sns.countplot('sex',data =df)
```

```python
sns.countplot(y='sex',data=df)
```
